```python
import pandas as pd 
import matplotlib.pyplot as plt

#  Goal 1.) NYC Water Consumption Per Capita over a 42 Year Period 
path = pd.read_csv('Water_Consumption_in_the_City_of_New_York copy.csv')
frame = pd.DataFrame(path)
x = frame['Year']
y = frame['Per Capita(Gallons per person per day)']
by_year_cap = frame[['Year','Per Capita(Gallons per person per day)']]
plt.plot(x, y)
plt.xlabel('Year')
plt.ylabel('Per Capita(Gallons per person per day)')
print(by_year_cap)
plt.show()
```

```python
#  Goal 2.)Averge total daily  NYC Water Consumption Over 42 year period 
path = pd.read_csv('Water_Consumption_in_the_City_of_New_York copy.csv')
frame = pd.DataFrame(path)
x = frame['Year']
y = frame['NYC Consumption(Million gallons per day)']
by_year_cap = frame[['Year','NYC Consumption(Million gallons per day)']]
plt.plot(x, y)
plt.xlabel('Year')
plt.ylabel('NYC Consumption(Million gallons per day')
print(by_year_cap)
plt.show()

```

```python
# Goal 3.) Aggerate NYC Population over 42 year period 
path = pd.read_csv('Water_Consumption_in_the_City_of_New_York copy.csv')
frame = pd.DataFrame(path)
x = frame['Year']
y = frame['New York City Population']
by_year_pop = frame[['Year','New York City Population' ]]
print(by_year_pop)
plt.plot(x,y)
plt.xlabel('Year')
plt.ylabel('New York City Population')
plt.show()
```
